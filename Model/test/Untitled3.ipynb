```python
def convert_str_to_nominal_variables(data2, col_name):
    # input: col_name(str)
    data_tmp = data2[col_name]
    cat = list(enumerate(list(data_tmp.unique())))
    table = dict()
    for item in cat:
        table[item[1]] = item[0]
    print (len(table), table )
    data2 = data2.replace({col_name: table })
    return data2
```
